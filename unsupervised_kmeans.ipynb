```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt
import pandas as pd

# Load inbuilt dataset
iris = load_iris()
X = iris.data
true_labels = iris.target  # Only used for comparison

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
predicted_labels = kmeans.labels_

# Evaluation metrics for clustering
print("\n=== Clustering Evaluation Metrics ===")
print("Inertia (WCSS):", kmeans.inertia_)
print("Silhouette Score:", silhouette_score(X_scaled, predicted_labels))
print("Calinski-Harabasz Score:", calinski_harabasz_score(X_scaled, predicted_labels))
print("Davies-Bouldin Score:", davies_bouldin_score(X_scaled, predicted_labels))

# Optional: Compare with true labels using a DataFrame
df = pd.DataFrame({
    "True Label": true_labels,
    "Cluster Label": predicted_labels
})
print("\nLabel Comparison (sample):\n", df.head())

# Optional: Visualize the clusters
plt.figure(figsize=(6, 4))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=predicted_labels, cmap='viridis', s=50)
plt.title("K-Means Clusters (Iris Dataset)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
